```python

import time
import subprocess
from getpass import getpass
from xvfbwrapper import Xvfb
import undetected_chromedriver as uc

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# ==============================================================================
# SET UP VIRTUAL DISPLAY
# This is crucial for running a browser and recording in a headless environment
# ==============================================================================
vdisplay = Xvfb()
vdisplay.start()

# ==============================================================================
# MAIN BOT FUNCTION WITH UNDETECTED CHROMEDRIVER
# ==============================================================================
def google_meet_bot(meet_link, google_email, google_password, recording_filename="meeting_recording.mp4"):
    """
    A Python bot to automatically join a Google Meet and record the screen using undetected ChromeDriver.

    Args:
        meet_link (str): The URL of the Google Meet.
        google_email (str): Your Google email address.
        google_password (str): Your Google password or app-specific password.
        recording_filename (str, optional): The filename for the screen recording.
                                            Defaults to "meeting_recording.mp4".
    """

    # ==========================================================================
    # UNDETECTED CHROMEDRIVER SETUP (Your recommended approach)
    # ==========================================================================
    options = uc.ChromeOptions()

    # Essential options for headless operation in Colab
    options.arguments.extend([
        "--no-sandbox",
        "--disable-setuid-sandbox",
        "--headless",
        "--disable-dev-shm-usage",
        "--window-size=1920,1080",
        "--disable-gpu",
        "--disable-extensions",
        "--disable-plugins",
        "--disable-images",
        "--disable-javascript",  # Disable JS for better performance
        "--mute-audio",
        "--disable-background-timer-throttling",
        "--disable-backgrounding-occluded-windows",
        "--disable-renderer-backgrounding"
    ])

    # Media permissions - block camera and microphone prompts
    prefs = {
        "profile.default_content_setting_values.media_stream_mic": 2,
        "profile.default_content_setting_values.media_stream_camera": 2,
        "profile.default_content_setting_values.geolocation": 2,
        "profile.default_content_setting_values.notifications": 2,
        "profile.managed_default_content_settings.javascript": 2
    }
    options.add_experimental_option("prefs", prefs)
    options.add_argument("--use-fake-ui-for-media-stream")

    # Initialize undetected ChromeDriver
    try:
        driver = uc.Chrome(options=options, version_main=105)  # Specify Chrome version
    except Exception as e:
        print(f"Failed to initialize with version 105, trying default: {e}")
        driver = uc.Chrome(options=options)

    wait = WebDriverWait(driver, 30)

    try:
        # --- Google Login ---
        print("Logging into Google Account...")
        driver.get("https://accounts.google.com/signin/v2/identifier?flowName=GlifWebSignIn&flowEntry=ServiceLogin")

        # Wait for email field and enter email
        email_field = wait.until(EC.visibility_of_element_located((By.NAME, "identifier")))
        email_field.clear()
        email_field.send_keys(google_email)

        next_button = wait.until(EC.element_to_be_clickable((By.ID, "identifierNext")))
        next_button.click()

        # Wait for password field and enter password
        password_field = wait.until(EC.visibility_of_element_located((By.NAME, "Passwd")))
        password_field.clear()
        password_field.send_keys(google_password)

        password_next = wait.until(EC.element_to_be_clickable((By.ID, "passwordNext")))
        password_next.click()

        print("Login Successful.")
        time.sleep(5)

        # --- Join Google Meet ---
        print(f"Joining Google Meet: {meet_link}")
        driver.get(meet_link)

        # --- Turn off mic and camera and join ---
        print("Preparing to join the meeting...")
        time.sleep(10)

        # Multiple strategies to turn off mic and camera
        try:
            # Strategy 1: Try to find and click mic/camera buttons
            mic_buttons = driver.find_elements(By.CSS_SELECTOR, '[data-is-muted="false"]')
            if mic_buttons:
                mic_buttons[0].click()
                print("Microphone turned off")
                time.sleep(1)

            camera_buttons = driver.find_elements(By.CSS_SELECTOR, '[data-is-camera-on="true"]')
            if camera_buttons:
                camera_buttons[0].click()
                print("Camera turned off")
                time.sleep(1)

        except Exception as e:
            print(f"Could not find buttons to turn off mic/camera: {e}")
            print("Proceeding to join with default settings.")

        # Try different selectors for join button
        join_selectors = [
            "//button[.//span[text()='Join now']]",
            "//button[contains(text(), 'Join now')]",
            "//button[contains(@aria-label, 'Join')]",
            "[data-tooltip='Join now']",
            ".uArJ5e.UQuaGc.Y5sE8d.VkkpIf.NqnKzc"
        ]

        joined = False
        for selector in join_selectors:
            try:
                if selector.startswith("//"):
                    join_button = wait.until(EC.element_to_be_clickable((By.XPATH, selector)))
                else:
                    join_button = wait.until(EC.element_to_be_clickable((By.CSS_SELECTOR, selector)))
                join_button.click()
                print("Successfully joined the meeting.")
                joined = True
                break
            except Exception as e:
                print(f"Join button selector failed: {selector} - {e}")
                continue

        if not joined:
            print("Could not find join button. Waiting and taking screenshot for debugging...")
            driver.save_screenshot("join_button_debug.png")
            time.sleep(5)

        time.sleep(5)

        # --- Start Screen Recording ---
        print("Starting screen recording...")
        resolution = "1920x1080"
        display_num = vdisplay.new_display
        ffmpeg_command = [
            'ffmpeg', '-y', '-f', 'x11grab',
            '-s', resolution, '-r', '25',
            '-i', f':{display_num}',
            '-c:v', 'libx264', '-preset', 'ultrafast',
            '-pix_fmt', 'yuv420p', recording_filename
        ]
        recording_process = subprocess.Popen(ffmpeg_command, stdout=subprocess.PIPE, stderr=subprocess.PIPE)
        print(f"Recording started. The recording will be saved as '{recording_filename}'.")

        # --- Wait for user to stop the script ---
        print("\nBot is running. The screen is being recorded.")
        print("To stop the recording and exit the meeting, stop this Colab cell.\n")

        # Monitor the meeting and provide status updates
        meeting_duration = 0
        while True:
            time.sleep(60)  # Check every minute
            meeting_duration += 1
            print(f"Meeting duration: {meeting_duration} minutes")

            # Optional: Check if still in meeting
            try:
                # Look for meeting indicators
                if "meet.google.com" not in driver.current_url:
                    print("No longer in meeting. Stopping recording...")
                    break
            except Exception as e:
                print(f"Error checking meeting status: {e}")

    except Exception as e:
        print(f"\nAn error occurred: {e}")
        driver.save_screenshot("error_screenshot.png")
        print("A screenshot 'error_screenshot.png' has been saved for debugging.")
        print("Please check the screenshot in the Colab file browser.")

    finally:
        if 'recording_process' in locals() and recording_process.poll() is None:
            print("Stopping screen recording...")
            recording_process.terminate()
            recording_process.wait()
            print("Recording stopped.")

        if 'driver' in locals():
            driver.quit()

        vdisplay.stop()
        print("Bot has been terminated.")

# ==============================================================================
# ALTERNATIVE FUNCTION WITH BETTER ERROR HANDLING
# ==============================================================================
def test_undetected_chrome():
    """Test function to verify undetected ChromeDriver works"""
    print("Testing undetected ChromeDriver setup...")

    options = uc.ChromeOptions()
    options.arguments.extend([
        "--no-sandbox",
        "--disable-setuid-sandbox",
        "--headless",
        "--disable-dev-shm-usage"
    ])

    try:
        driver = uc.Chrome(options=options)
        driver.get("https://google.com")
        print("✅ Undetected ChromeDriver test successful!")
        print(f"Page title: {driver.title}")
        driver.quit()
        return True
    except Exception as e:
        print(f"❌ Undetected ChromeDriver test failed: {e}")
        return False

if __name__ == "__main__":
    # Test the setup first
    if test_undetected_chrome():
        print("\n" + "="*50)
        print("STARTING GOOGLE MEET BOT")
        print("="*50)

        # --- User Inputs ---
        MEET_LINK = input("Enter the Google Meet link: ")
        GOOGLE_EMAIL = input("Enter your Google email: ")
        GOOGLE_PASSWORD = getpass("Enter your Google password or app-specific password: ")

        google_meet_bot(MEET_LINK, GOOGLE_EMAIL, GOOGLE_PASSWORD)
    else:
        print("Setup test failed. Please check your Chrome and ChromeDriver installation.")
```
